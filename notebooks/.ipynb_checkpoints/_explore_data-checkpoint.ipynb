```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.grid'] = True
```

## Load data


```python
data_path = '../data/traffic_synthetic.csv'

df = pd.read_csv(data_path)
print(f"Shape: {df.shape}")
df.head()
```

## Basic info and summary statistics


```python
df.info()
df.describe(include='all').T
```

## Histograms for numerical features


```python
num_cols = df.select_dtypes(include=[np.number]).columns
print('Numerical columns:', list(num_cols))

if len(num_cols) == 0:
    print("No numerical columns found.")
else:
    df[num_cols].hist(bins=30, figsize=(15, 10), layout=None)
    plt.tight_layout()
    plt.show()
```

## Correlation matrix and heatmap


```python
if len(num_cols) > 1:
    corr = df[num_cols].corr()
    display(corr)

    fig, ax = plt.subplots()
    cax = ax.matshow(corr, interpolation='nearest')
    fig.colorbar(cax)

    ax.set_xticks(range(len(num_cols)))
    ax.set_yticks(range(len(num_cols)))
    ax.set_xticklabels(num_cols, rotation=90)
    ax.set_yticklabels(num_cols)

    ax.set_title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
else:
    print("Need at least two numeric columns for correlation.")
```

## Pairwise scatter plots (scatter matrix)


```python
if len(num_cols) > 1:
    max_cols = 6                     # keep readable
    cols_for_pairs = num_cols[:max_cols]

    print("Scatter matrix columns:", list(cols_for_pairs))
    
    scatter_matrix(df[cols_for_pairs],
                    figsize=(12, 12),
                    diagonal="hist")

    plt.suptitle("Scatter Matrix of Numerical Features", y=0.9)
    plt.show()
else:
    print("Need at least two numeric columns for scatter matrix.")
```

## Time-series traffic trends


```python
# Detect possible time column automatically
time_col_candidates = [
    c for c in df.columns
    if "time" in c.lower() or "date" in c.lower() or "timestamp" in c.lower()
]

time_col = None

for c in time_col_candidates:
    try:
        pd.to_datetime(df[c].iloc[:50], errors="raise")
        time_col = c
        break
    except Exception:
        continue

print("Detected time column:", time_col)

if time_col is None:
    print("No suitable time column detected. Set manually if required.")

else:
    df_ts = df.copy()
    df_ts[time_col] = pd.to_datetime(df_ts[time_col])
    df_ts = df_ts.sort_values(time_col)

    # heuristic to detect traffic related column
    traffic_col_candidates = [
        c for c in df_ts.select_dtypes(include=[np.number]).columns
        if any(k in c.lower() for k in
               ["traffic", "volume", "count", "flow", "vehicles"])
    ]

    if len(traffic_col_candidates) == 0:
        print("No explicit traffic column name found  using first numeric column.")
        traffic_col = num_cols[0] if len(num_cols) else None
    else:
        traffic_col = traffic_col_candidates[0]

    if traffic_col:
        print("Using traffic column:", traffic_col)

        plt.figure()
        plt.plot(df_ts[time_col], df_ts[traffic_col])
        plt.xlabel(time_col)
        plt.ylabel(traffic_col)
        plt.title(f"Time-Series Trend of {traffic_col}")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```
